With the function, df_we['temp_min'].plot(), what we say is: in the dataframe df_we, select the column temp_min and plot() the data. In the representation that Pandas has made, he has put on the x-axis the indices (numbers from 0 to 1460). We can give the plot() function more information so that it represents more precisely what we are interested in. In the next cell we explicitly say which data we represent on the x-axis are the dates and on the y-axis the maximum temperatures.

```python
df_we.plot(x='date', y='temp_min')
```

Increase the __size__ of the graph

```python
df_we.plot(x='date', y='temp_min',figsize=(10,5))
```

 We can paint __several columns__ of data on the same chart:

```python
df_we.plot(x='date', y=['temp_min','temp_max'], figsize=(10,5))
```

Add __title__, __labels__, __axes__ and __colour__:

```python
ax = df_we.plot(x='date', y=['temp_min','temp_max'], figsize=(10,5),
              title='Temperatures in Seattle',
              color=['lightgreen','red'])
ax.set(xlabel='date', ylabel='º Celsius')
```

```python
ax = df_we.plot(x='date', y=['temp_min','temp_max'], figsize=(10,5),
              title='Temperatures in Seattle',
              color=['deeppink','plum'])
ax.set(xlabel='date', ylabel='º Celsius')
```

It is also interesting to add __composite graphs__ that combine information from several variables.

```python
df_we.plot(x='date', y=['temp_min','temp_max', 'wind'], figsize=(10,5))
```

In the above graph there is only one scale and we are interested in creating a different scale for the wind variable, which is in metres per second, while the temperature is in degrees Celsius.

```python
df_we.plot(x='date', y=['temp_min','temp_max', 'wind'], figsize=(10,5), secondary_y='wind')
```

One way to improve the graphic is to use __transparency__ in the colours. This can be done by adding the parameter __alpha__ with a value between 0 and 1. A value of 1 means full opacity and a value of 0 means full transparency.

```python
df_we.plot(x='date', y=['temp_min','temp_max', 'wind'], figsize=(10,5), secondary_y='wind', alpha = 0.5)
```

We can also customise the axis limits with __ylim__.

```python
ax = df_we.plot(x='date', y=['temp_min','temp_max', 'wind'],
           figsize=(10,5), secondary_y='wind',
           alpha = 0.5, ylim = (-5, 35))
ax.right_ax.set_ylim(0,14) # eje de la derecha
```

Adding the labels

```python
ax = df_we.plot(x='date', y=['temp_min','temp_max', 'wind'],
           figsize=(10,5), secondary_y='wind',
           alpha = 0.5, ylim = (-5, 35), ylabel = "º Celsius")
ax.right_ax.set_ylim(0,14)
ax.right_ax.set_ylabel("Wind speed m/s")
```

Or contextualise it with a map from [openstreet maps](https://www.openstreetmap.org/#map=7/40.007/-2.488), *the wikipedia of maps*. For that one detail is needed, which is to indicate the projection we want. As you know, the Earth is (almost) a sphere. In order to make a 2D map, we need to perform an operation called [*projection*](https://es.wikipedia.org/wiki/Proyecci%C3%B3n_cartogr%C3%A1fica). The graphics used by contextily are in the so-called [Mercator projection](https://es.wikipedia.org/wiki/Proyecci%C3%B3n_de_Mercator). First we need to tell dataframe `madrid` that it is going to use that projection.

```python
madrid = madrid.to_crs(epsg=3857) # This is necessary to contextualise the map. contextily uses by default
                                  # the mercator projection.
ax = madrid.plot(figsize=(14,14), alpha=0.5)
ctx.add_basemap(ax, zoom=12)
```
